### Arithematic Operations in NumPy

### 1D Array

Addition in Numpy - **np.add(var1,var2)**

```python
var1 = np.array([1,2,3,4])
var2 = np.array([5,6,7,8])

varad = var1+3
varadd = np.add(var1,3) #Both are same
print(varadd)

varadd1 = var1 + var2 # Method 1 for addition
print(varadd1)

varadd2 = np.add(var1,var2) #Method 2
print(varadd2)
```

```python
var1 = np.array([1,2,3,4])
var2 = np.array([5,6,7,8])

varad = var1-3
varadd = np.subtract(var1,3) #Both are same
print(varadd)

varadd1 = var1 - var2 # Method 1 for subtraction
print(varadd1)

varadd2 = np.subtract(var1,var2) #Method 2
print(varadd2)
```

Multiplication- **.multiply(var1,var2)**

```python
var1 = np.array([1,2,3,4])
var2 = np.array([5,6,7,8])

varad = var1*3
varadd = np.multiply(var1,3) #Both are same
print(varadd)

varadd1 = var1 * var2 # Method 1 
print(varadd1)

varadd2 = np.multiply(var1,var2) #Method 2
print(varadd2)
```

Division- **.divide(var1,var2)**

```python
var1 = np.array([1,2,3,4])
var2 = np.array([5,6,7,8])

varad = var1/3
varadd = np.divide(var1,3) #Both are same
print(varadd)

varadd1 = var1 / var2 # Method 1 
print(varadd1)

varadd2 = np.divide(var1,var2) #Method 2
print(varadd2)
```

Modulus - **.mod(var1,var2)**

```python
var1 = np.array([1,2,3,4])
var2 = np.array([5,6,7,8])

varadd1 = var1 % var2 # Method 1
print(varadd1)

varadd2 = np.mod(var1,var2) #Method 2
print(varadd2)
```

Power - **.power(var1,var2)**

```python
var1 = np.array([1,2,3,4])
var2 = np.array([5,6,7,8])

varadd1 = var1 ** var2 # Method 1
print(varadd1)

varadd2 = np.power(var1,var2) #Method 2
print(varadd2)
```

Reciprocal- **.reciprocal(var1)**

```python
var1 = np.array([1,2,3,4])


varadd1 = 1/var1 # Method 1
print(varadd1)

varadd2 = np.reciprocal(var1, dtype='f') #Method 2
print(varadd2)
```

## Arithmetic Functions

**.max(X)** - Maximum term in the array                               
**.min(X)** - Minimum Term in the array                               
**.agrmin(X)** - Position of the min. term                                   
**.argmax(X)** - Position of the max. term                                    

```python
# 1D Array
var = np.array([4,5,32,1,5,7,5,3])

print(f"Min: {np.min(var)} & position: {np.argmin(var)}")
print(f"Max: {np.max(var)} & position: {np.argmax(var)}")

#2D Array
# axis = 0 -> column kai respect mai
# axis = 1 -> row kai respect mai 
var1 = np.array([[2,3,1],[4,7,2]])

print(f"Min: {np.min(var1,axis=1)} & Position: {np.argmin(var1,axis=1)}") #Row-wise min
print(f"Max: {np.max(var1,axis=0)} & Position: {np.argmax(var1,axis=0)}") #Column wise max
```

### Broadcasting in NumPy Arrays

Broadcasting in NumPy allows arrays of different dimensions to be used together in arithematic operations by automatically expanding the smaller array to match the shape of larger array

**How it Works -**
When performing operations (like addition, subtraction, multiplication,etc) on 2 arrays, Numpy compares there shape element-wise. If their shapes is not same,Numpy tries to broadcast smaller array to make their shapes compatible

**Rules for Broadcasting -**
1) **Align Dimensions from Right -** When comparing the shapes of 2 arrays, align their dimensions from the rightmost (trailing) side.
2) **Dimensions are compatible if -** 
**First-** The dimensions are equal
**Second-** One of the dimensions is 1

**Steps for Broadcasting -**
1) Start comparing the shapes of arrays from the rightmost side
2) If the dimensions are same, keeps it as-it is
3) If one dimension is 1, "Expand" it to match the other dimension
4) If neither dimension is 1 or equal , Broadcasting Fails 

```python
# Arrays with Compatible shapes
arr1 = np.array([[1,2,3],[4,5,6]])  #Shape (2,3)

arr2 = np.array([10,20,30]) #Shape (3,)
# Broadcasting -  arr2 becomes [[10,20,30],[10,20,30]] -> Shape (2,3)

print(arr1+arr2)
```

**Q Why is the shape of arr2 (3,) instead of (1,3) -**
**Reason -** A shape of (3,) means the array is 1-D with 3 elements. There is no explicit concept of rows and columns in 1-D Array

```python
# Arrays with Different Dimensions
arr1 = np.array([[1],[2],[3]])  # Shape (3,1)
arr2 = np.array([10,20,30])  # Shapr (3,)
# Broadcasting arr2 - arr2=[[10,20,30],[10,20,30],[10,20,30]] (Image mai working hai)
print(arr1+arr2)
```

```python
# When Broadcasting Fails(ValueError) -  If shapes are not compatible
# Type 1
arr1 = np.array([1,2])  #Shape (2,)
arr2 = np.array([1,2,3]) #Shape (3,)
print(arr1+arr2)

# Type 2
arr1 = np.array([[2,1],[3,1]])  #Shape (2,2)
arr2 = np.array([[1,2,3],[1,2,3]]) #Shape (2,3)
print(arr1+arr2)
```
