### **Bar Plots**
- used to visualize categorical data with rectangular bars. 
- The length of each bar represents the value of the category.

**Syntax -**
plt.bar(categories, values, color='blue', width=0.8, edgecolor='black', linewidth=1)

**Note- Default Vertical Bar Graph**

- categories: List of category names (x-axis)
- values: Corresponding values for each category (y-axis)
- color: Changes bar color
- width: Adjusts bar thickness (default is 0.8)
- edgecolor: Sets the border color of bars
- linewidth: Thickness of the border
- linestyle: Changes the border line style
- alpha: Colour density increase ya decrease karne kai liye
- rotation: Labels ko tilt karne kai liye

```python
# Basic Bar Plot
x = ["A",'B','C',"D"]
y = [20,35,30,40]
c = ["blue","red","black","green"]
plt.bar(x, y, color=c, width=0.4,edgecolor="black", linewidth=2, linestyle="--", alpha=0.8,label="class 9")
plt.xlabel('Sections',fontsize = 15)
plt.ylabel('No. of Students', fontsize=15)
plt.title('Basic Bar Plot example')
plt.legend()
plt.show()
```

**##### Horizontal Bar Chart**
- Same functions as before but vertical kai jagah horizontal

**Syntax - plt.barh(x,y,etc)**

**Note- barh does not accept width parameter, instead it accepts height paramter**

```python
x = ["A",'B','C',"D"]
y = [20,35,30,40]
c = ["blue","red","black","green"]
plt.barh(x, y, color=c,height=0.4, edgecolor="black", linewidth=2, linestyle="--", alpha=0.8,label="class 9")
plt.xlabel('Sections',fontsize = 15)
plt.ylabel('No. of Students', fontsize=15)
plt.title('Basic Bar Plot example')
plt.legend()
plt.show()
```

##### **Grouped Bar Chart**
- Used when comparing multiple sets of data

```python
x = ["A","B","C","D"]
values1 = [20,35,30,40]
values2 = [25,32,34,20]

p = np.arange(len(x))  # Array create karke indexing create karte hai for x-axis

plt.bar(p - 0.2, values1,width=0.4, label="Class 10", color='blue')
plt.bar(p + 0.2, values2,width=0.4, label="Class 9", color='green')

plt.xticks(p,x,rotation=10) #Maps numeric positions back in category labels
plt.xlabel("Sections")
plt.ylabel("No. of students")
plt.title("Comparing Section wise division")
plt.legend()
plt.show()
```

##### **Stacked Bar Chart**

```python
x = ["A","B","C","D"]
values1 = [20,35,30,40]
values2 = [25,32,34,20]

plt.bar(x, values1,width=0.4, label="Class 10", color='blue')
plt.bar(x, values2,width=0.4, bottom=values1, label="Class 9", color='green')

plt.xlabel("Sections")
plt.ylabel("No. of students")
plt.title("Stacked Bar Chart")
plt.legend()
plt.show()
```

#### **Scatter Plot**
- used to visualize the relationship b/w 2 variables (numerical data) as points in 2D space. 
- ideal for showing patterns, trends and outliers in the data.


- marker - changes the shape of the points ("o","x","^","s")

```python
# Basic Scatter Plot
x = [1,2,3,4,5]
y = [10,20,25,30,40]
sizes = [50,100,200,300,400]
colors = ['red','blue','green',"orange","purple"]

plt.scatter(x,y, c=colors,s=sizes, label="Dataset 1",marker="s",alpha= 0.8, edgecolor='black',linewidth=1.5)
plt.xlabel("X-Axis Label")
plt.ylabel("Y-Axis Label")
plt.title("Basic Scatter Plot")
plt.legend()
plt.grid(False) #Shows grid if True
plt.show()
```

##### **ColourMap for Continuous Data (cmap)**
- Jab nos. pass karte hai colors mai tab cmap ka use karte hai

```python
z = np.random.rand(5)

plt.scatter(x,y,c=z, cmap='viridis',s=100)
plt.colorbar(label='Color Scale') #Adds a color scale legend
plt.show()
```

##### **Multiple Scatter Plots**

```python
x1 = x2 = [1,2,3,4,5]
y1 = [10,20,25,30,40]
y2 = [15,25,20,35,45]

plt.scatter(x1,y1, c='blue', s=100, label="Dataset 1")
plt.scatter(x2, y2, c='red', s=100, label="Dataset 2")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Multiple Scatter Plots")
plt.legend()
plt.show()
```

##### **Annotating Points in Scatter Plots**

```python
points = ['A', 'B', 'C', 'D', 'E']

for i in range(len(x)):  # Adds text 0.1 unit away from scatter plot
    plt.text(x[i] + 0.1, y[i] + 0.1, points[i], fontsize=9) 

plt.scatter(x, y, color='blue', s=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Annotations')
plt.show()
```

#### **Line Plot -**
- Basic chart type used to visualize trends, relationships or continuous data over time.
- It connects data points with a line to display patterns or changes.

**Syntax - plt.plot(x, y, linestyle=, color=, marker=, label=,etc)**



```python
# Basic Line Plot

x=[1,2,3,4,5]
y=[10,20,25,30,40]

for i in range(len(x)): #Annotating on the points
    plt.text(x[i]-0.4, y[i]+0.7, f'({x[i],y[i]})', fontsize=9)

plt.plot(x, y, label="Dataset 1", color='blue', linestyle="--", marker="o", markersize=5,linewidth=2.5)
plt.xlabel('X-axis Label')
plt.xticks(x)
plt.ylabel('Y-axis Label')
plt.title('Basic Line Plot with Annotations')
plt.legend()
plt.show()
```

##### **Plotting Multiple Lines**

```python
x = [1,2,3,4,5]
y1 = [10,20,25,30,40]
y2 = [15,25,20,35,45]

plt.plot(x, y1, c="blue",label="Dataset 1",marker="s")
plt.plot(x, y2, c="red", label="Dataset 2",marker="o")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plots')
plt.legend()
plt.show()
```

##### **Fill Area under the Line**
- we use plt.fill_between()

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, c="purple", label="Sine Wave")
plt.fill_between(x, y, color="green",alpha=0.4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Shaded Area')
plt.legend()
plt.show()
```

##### Real-Life Eg - Stock Price Trends

#### **Pie Chart-**
- used to visualize proportions and percentages of different categories in a dataset.
- Displays data as slices of a circle, where the size of each slice represents the proportion of that category.




- pctdistance=1.2 - Moves the percentages values outside

```python
# Basic Pie Chart Example
labels = ['Python','Java','C++','JavaScript']
sizes = [40,25,20,15]  # Percentages
colors = ['blue','red','green','yellow']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.2f%%', startangle=140,shadow=True,explode=[0.05,0,0,0])
plt.title('Programming Language Popularity')
plt.show()
```

##### **Multiple Pie Charts (subplots)**
- you can display multiple charts in a single fig. using plt.subplots().

Syntax - fig, ax = plt.subplots(rows, cols, figsize=(width, height))
ax[index].pie(sizes, labels, autopct, etc.)


```python
labels = ['Clothing', 'Electronics','Groceries','Others']
sales_2023 = [40,25,20,15]
sales_2024 = [35,30,25,10]

fig, ax = plt.subplots(1,2, figsize=(12,6))

ax[0].pie(sales_2023, labels=labels, autopct='%1.1f%%', startangle=90, colors=['blue','red','green','orange'])
ax[0].set_title('Sales Distribution - 2023')

ax[1].pie(sales_2024, labels=labels, autopct='%1.1f%%', startangle=90, colors=['blue', 'red', 'green', 'orange'])
ax[1].set_title('Sales Distribution - 2024')

plt.show()
```

#### **Stem Plot**
- also known as Lollipop Chart , is a type of plot that represents discrete data by drawing vertical lines(stems) from a baseline to each data point with a marker at the top. 
- It's useful for visualizing individual values while maintaining a clean, simple representation.



- **use_line_collection = true ->** Improves rendering for large datasets
- **basefmt=' ' →** Hides the baseline.

```python
# Basics Stem Plot 
x = np.array([1,2,3,4,5])
y = np.array([10,15,7,25,20])

plt.stem(x, y, linefmt='b-', markerfmt='rx', basefmt='k--')

for i,value in enumerate(y):   # Annotating in Stem Plot
    plt.text(x[i], y[i]+1, str(value), ha='center',  fontsize=10)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Stem Plot')
plt.show()
```

##### **Multiple Stem Plots**


```python
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([10, 15, 7, 25, 20])
y2 = np.array([5, 10, 12, 18, 22])

plt.stem(x, y1, linefmt='b-', markerfmt='bo', basefmt='k--', label="Dataset 1")
plt.stem(x, y2, linefmt='r-', markerfmt='ro', basefmt='o--', label='Dataset 2')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Stem Plots')
plt.legend()
plt.show()
```
